## Data Cleaning 

```python
ec['type'] = ec['type'].fillna(ec['type'].mode()[0])
```

```python
ec[ec["Turnout"].isnull()]
```

```python
ec[ec["year"].isnull()]
```

```python
ec = ec.dropna(axis=0, subset=['Turnout', 'margin%', 'year'])
```

```python
ec.isnull().sum()
```

## Remove unwanted signs 

```python
ec['Turnout'] = ec['Turnout'].str.replace('%', '')
ec['margin%'] = ec['margin%'].str.replace('%', '')
```

```python
ec[ec["Turnout"] == '-']
```

```python
ec = ec.drop(ec[ec['Turnout'] == '-'].index)
```

```python
ec.shape
```

```python
ec['electors'] = ec.electors.apply(lambda x: x.strip(','))

ec['electors'] = ec.electors.apply(lambda x: x.replace(',',''))
```

```python
ec['votes'] = ec.votes.apply(lambda x: x.strip(','))

ec['votes'] = ec.votes.apply(lambda x: x.replace(',',''))
```

```python
ec['margin'] = ec.margin.apply(lambda x: x.strip(','))

ec['margin'] = ec.margin.apply(lambda x: x.replace(',',''))
```

```python
ec.head()
```

## Change Datatypes 

```python
ec['no'] = ec['no'].astype(int)
ec['electors'] = ec['electors'].astype(int)
ec['votes'] = ec['votes'].astype(int)
ec['Turnout'] = ec['Turnout'].astype(float)
ec['margin'] = ec['margin'].astype(int)
ec['margin%'] = ec['margin%'].astype(float)
ec['year'] = ec['year'].astype(int)
```

```python
ec = ec.drop(['Pc_name', 'no', 'type', 'candidate_name'], axis =1)
```

```python
ec_num = ec.select_dtypes(include =['int32','float64'])
ec_num.head()
```

```python
ec_cat = ec.select_dtypes(include = ['object'])
ec_cat.head()
```

```python
ec['turnout'] = ec.apply(lambda x: '1' if x['Turnout'] >= 55 else '0', axis=1)
```

```python
ec.sample(5)
```

```python
ec.turnout.value_counts()
```

```python
ec = ec.drop(['Turnout'], axis=1)
```

## One Hot Encoding 

```python
ec = pd.get_dummies(ec, columns=['state','party'], drop_first=True)
ec.head()
```

```python
X = ec.loc[:, ec.columns != 'turnout']
X.head()
```

```python
y = ec.turnout
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
numericfeatures = list(X_train.columns)[ :5]
numericfeatures
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train[numericfeatures] = ss.fit_transform(X_train[numericfeatures])

X_test[numericfeatures] = ss.transform(X_test[numericfeatures])
```

```python
X_train.head()
```

```python
# Modeling

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Prediction

y_pred_train = logreg.predict(X_train)
y_pred_test = logreg.predict(X_test)
```

```python
y_pred_train
```

```python
y_train
```

```python
# Model Evaluation

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy_score(y_test, y_pred_test)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
cm = pd.DataFrame(confusion_matrix(y_test, y_pred_test))
cm
```
